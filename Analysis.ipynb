```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("results.csv") 
df
```

```python
df['Company'].unique()
```

## LINE GRPAH

```python
sns.set_style('whitegrid')
plt.figure(figsize=(15,10))
sns.lineplot(data=df, x="Hour", y="High", hue="Company")
plt.xlabel("Hour",fontsize=15)
plt.ylabel("Stock Price",fontsize=15)
plt.title("Movement of Stock Prices",fontsize=18)
```

## Opening vs Closing Price

```python
sns.set(style="whitegrid")

df9 = df[(df['Hour'] == 9)]

fig = plt.figure(figsize=(15,5))

bar_company = sns.barplot(x="Company", y="High", data=df9)

bar_company.set(xlabel='Company', ylabel='Highest Price at first trading hour')

plt.show()
```

```python
sns.set(style="whitegrid")

df9 = df[(df['Hour'] == 15)]

fig = plt.figure(figsize=(15,5))

bar_company = sns.barplot(x="Company", y="High", data=df9)

bar_company.set(xlabel='Company', ylabel='Highest Price at first trading hour')

plt.show()
```

## LINE GRAPH TO TRACK UPWARD AND DOWNWARD MOVEMENT OF STOCK PRICES

```python
for x in df['Company'].unique():
    print('Company:',x)
    line = df[df['Company'] == x].loc[:,['Hour','High']]
    print(line.set_index('Hour').plot(kind='line'))
    plt.show()
```
